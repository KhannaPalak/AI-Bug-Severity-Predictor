```python
import pandas as pd

test_df = pd.read_csv("../../data/processed/test.csv")

print(test_df.head())
```

```python
test_df = test_df.dropna(subset=["text", "target"])

print(len(test_df))
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("../../models/bert_model")
model = AutoModelForSequenceClassification.from_pretrained("../../models/bert_model")
```

```python
from datasets import Dataset

test_dataset = Dataset.from_pandas(test_df)

def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        padding="max_length",
        max_length=256
    )

test_dataset = test_dataset.map(tokenize, batched=True)
```

```python
test_dataset = test_dataset.rename_column("target", "labels")

test_dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)
```

```python
from transformers import Trainer

trainer = Trainer(model=model)
```

```python
import torch

print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")

print(next(model.parameters()).device)
```

```python
trainer.args.per_device_eval_batch_size
```

```python
from transformers import TrainingArguments

eval_args = TrainingArguments(
    output_dir="./temp_eval",
    per_device_eval_batch_size=64
)

trainer = Trainer(
    model=model,
    args=eval_args
)
```

```python
predictions = trainer.predict(test_dataset)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

bert_pred = np.argmax(predictions.predictions, axis=1)

bert_accuracy = accuracy_score(test_df["target"], bert_pred)

bert_precision = precision_score(
    test_df["target"],
    bert_pred,
    average="weighted",
    zero_division=0
)

bert_recall = recall_score(
    test_df["target"],
    bert_pred,
    average="weighted",
    zero_division=0
)

bert_f1 = f1_score(
    test_df["target"],
    bert_pred,
    average="weighted",
    zero_division=0
)
```

```python
print("Accuracy :", bert_accuracy)
print("Precision:", bert_precision)
print("Recall   :", bert_recall)
print("F1 Score :", bert_f1)
```
